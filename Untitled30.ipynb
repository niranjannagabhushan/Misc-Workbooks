```python
vowelcount = 0;  
consocount = 0;  
digit = 0;
str = "This is Edubridge Course";  

str = str.lower();
for i in range(0,len(str)):   
    #Checks whether a character is a vowel  
    if str[i] in ('a',"e","i","o","u"):  
        vowelcount = vowelcount + 1;  
    elif (str[i] >= 'a' and str[i] <= 'z'):  
        consocount = consocount + 1;
    else (str[i] >= '0' and str[i] <= '9' ]):
        digit = digit + 1:

print("Total number of vowel and consonant are" );  
print(vowelcount);  
print(consocount);
print(digit);
```
