### **Convolutional Neural Nets (ConvNets or CNN) are a class of neural network specialized for image processing. Like other neural networks, they transform input to output through many layers. In CNN layers have a Convolution step, a pooling step (optional) and a non-linear activation. Each layer in a neural network transforms the input tensor into the output tensor through linear and non-linear operations. All these intermediate tensors (including the network input and output) are called activations and they are all different representations of the input.**

![](https://miro.medium.com/max/1250/1*F2Ik_XFzmu5jZF-byiAKQQ.jpeg)

### **In year 1981, the Nobel Prize in Medicine was awarded to Hubel and Wiesel for their fundamental research for the understanding of Mammalian (Cat, Monkeys, etc.) brains to understand how the Neuronal system of Vision works. They focus mainly on different kinds of Visual tasks.**


![](https://i.imgur.com/M6nUOFV.png)


![](https://miro.medium.com/max/1400/1*YeiElg3W7wDosp38lwyUig.png)

### **So, as we can see in the above picture they have taken a Cat and gave him anesthesia so the cat can’t move. But his brain and other body parts are working absolutely fine. They show a Light Bar on a Screen as shown in the picture, and the researchers attached a small probe, which is called Micro-Electrode. They attached this to the brain of the cat, especially to the Visual Cortex, and started recording. And they observe that for some specific types of edges or light patterns the Neurons (Electrical pulses) behave differently than others.**

### **As shown in the diagram, for the vertical pattern there is a lot of electron firing, and for the Horizontal pattern, there are no electron firings. As we move from Horizontal to Vertical it starts to fire more and more electrons. So from this experiment, they concluded that there must be some special neurons in the cat’s brain which are firing when they see a Vertical edge.**


![](https://i.imgur.com/chFzd8O.png)

### **Here (n x n) is the input, (k x k) is the convolution and (n-k+1)(n-k+1) is the output.**
### **Now let’s say, we have a 6 x 6 (n = 6) matrix and a 3 x 3 kernel (K = 3) and So we will get the output as a 4 x 4 matrix.**
### **But, what if we don't want to shrink it and want to have the same output matrix as the input matrix.**
### **To overcome this problem we will apply a trick call Padding.**

![](https://miro.medium.com/max/875/1*4YRzHEi414VH0ECsd-qPpg.png)

### **Here we are doing padding on the given 6 x 6 matrix and making it an 8 x 8 input matrix. Because we know that if we want to get a 6 x 6 output matrix we will have to take an 8 x 8 matrix.**
### **Here we are doing P = 1 and fill the padding with zeros (0). This is called Zero Padding.**
### **Instead of filling it with Zeros (0), we can fill it with the nearest values from the cells. This is called the Same Value Padding.**
### **There are other types of padding also like Raw Padding, Column Padding, Row Padding etc.**


![](https://i.imgur.com/r9u9Mdy.png)


![](https://miro.medium.com/max/875/1*VTANvoVFMXvc-aKHkMqcAg.png)

### **Previously, we are placing the 3 x 3 kernel matrix on top of the input 6 x 6 matrix and doing componentwise multiplication and then addition. Then we are shifting 1 cell. This is called Stride by 1.**
### **Similarly, we can also Shift 2 cells by simply doing Stride = 2 (s = Stride Length)**
### **When we have Strides length of 2 it will simply shift 2 cells both horizontally and vertically.**

![](https://miro.medium.com/max/875/1*4lmzygoiiC7EmunzUBC-nw.png)

![](https://miro.medium.com/max/875/1*8u9VRtYJpiVCu47cvTq4lw.png)

### **Deep Neural Networks tend to Overfit due to many weights. To avoid this overfitting problem we can use a technique called Dropout. Dropout is an extremely simple yet very elegant technique.**

### **So in the diagram below, we can clearly see we are using x1, x2, x3, and x4, this is the input layer, and L1 is 1st hidden layer and L2 is 2nd hidden layer and there is our Output layer.**

### **So the basic idea of Dropouts are, in any given iteration in any given layer (can be input layer or hidden layers) and randomly take a subset of the inputs and remove all the connections that are incoming or outgoing from that particular input.**

### **Fully Connected Layers are nothing but Feed-Forward Networks. After the pooling layer, we flattened our data into a vector and then feed that into a Fully Connected Layer just like in a simple Neural Network.**

### **So let’s understand this with an example, let’s say, if we are detecting a dog, the Convolutional layers may be able to detect the different features like eyes, nose, ears, tongue, etc. But the problem is, they will not be able to get the exact positions of these features. So to combat that, we will use Fully Connected Layers, we will combine these features together will create a more complex model to able to detect the Dogs properly.**


![](https://i.imgur.com/FAPSv63.png)


### **As the human brain never stops learning, the brain always learns from the observations and experiences i.e. the inputs which it receives from the sensory organs are utilized at some or another point but the learning never becomes “Zero”.**
### **To add this feature to the neural networks ReLU is used. The activation function is : f(x) = max(0,x). For any activation function, we must be able to take the derivative of that function and with ReLU we can do that. But the derivative at zero is not defined for the ReLU. Due to zero, we can have the problem of a dead activation state. This implies there will be no weight change meaning no learning. But for humans, it doesn’t happen very often. To solve this problem the concept of Leaky ReLU is introduced.**


![](https://i.imgur.com/eg1uEJH.png)
