```python
df1 = df[df['hits']>0]
df2 = df1[df1['clicks']>0]
df3 = df2[df2['clickThroughs']>0]
df4 = df3[df3['ctr']>0]
df5 = df4[df4['carts']>0]
df6 = df5[df5['orders']>0]
```
