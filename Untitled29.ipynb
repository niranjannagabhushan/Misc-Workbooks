```python
print("Enter Four Grades: ")
m1 = int(input())
m2 = int(input())
m3 = int(input())
m4 = int(input())

total = (m1+m2+m3+m4)
avg = total/4

if avg>=91 and avg<=100:
  print("A")
elif avg>=81 and avg<=91:
  print("0")
elif avg>=71 and avg<=81:
  print("B")
elif avg>=51 and avg<=61:
  print("C")
else:
  print("invalid")
```

```python
# odd
num = int(input("N : "))
sum = 0;

for i in range(1, num + 1):

    #Check for odd or not.
    if( (i % 2) != 0):
        sum += i;

print("\nSum of odd numbers from 1 to", num, "is :", sum)
```

```python
# even
num = int(input("N : "))
sum = 0;

for i in range(1, num + 1):

    #Check for even or not.
    if( (i % 2) == 0):
        sum += i;

print("\nSum of even numbers from 1 to", num, "is :", sum)
```
